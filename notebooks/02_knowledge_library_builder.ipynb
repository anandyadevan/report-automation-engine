```python
from openpyxl import Workbook

wb = Workbook()

# ==========================================================
# Metric Library
# ==========================================================
ws = wb.active
ws.title = "Metric Library"

ws.append([
    "variable",
    "level",
    "label",
    "color",
    "description"
])

for variable, levels in METRIC_LIBRARY.items():

    if variable == "grade":

        for grade, info in levels.items():

            ws.append([
                variable,
                grade,
                info["label"],
                info["color"],
                info["description"]
            ])

    else:

        for level, info in levels.items():

            if level == "name":
                continue

            ws.append([
                variable,
                level,
                info["label"],
                info["color"],
                info["description"]
            ])


# ==========================================================
# Interpretation Library
# ==========================================================
ws = wb.create_sheet("Interpretation Library")

ws.append([
    "variable",
    "level",
    "title",
    "description"
])

for variable, levels in INTERPRETATION_LIBRARY.items():

    for level, info in levels.items():

        ws.append([
            variable,
            level,
            info["title"],
            info["description"]
        ])


# ==========================================================
# Subject Interpretation
# ==========================================================
ws = wb.create_sheet("Subject Interpretation")

ws.append([
    "variable",
    "level",
    "title",
    "description"
])

for variable, levels in SUBJECT_INTERPRETATION_LIBRARY.items():

    for level, info in levels.items():

        ws.append([
            variable,
            level,
            info["title"],
            info["description"]
        ])


# ==========================================================
# Learning Plan
# ==========================================================
ws = wb.create_sheet("Learning Plan")

ws.append([
    "variable",
    "level",
    "goal",
    "actions",
    "timeline",
    "success_indicator"
])

for variable, levels in LEARNING_PLAN_LIBRARY.items():

    for level, info in levels.items():

        ws.append([
            variable,
            level,
            info["goal"],
            " | ".join(info["actions"]),
            info["timeline"],
            info["success_indicator"]
        ])


# ==========================================================
# Growth Opportunity
# ==========================================================
ws = wb.create_sheet("Growth Opportunity")

ws.append([
    "variable",
    "threshold",
    "title",
    "description"
])

for variable, levels in GROWTH_OPPORTUNITY_LIBRARY.items():

    info = levels["very_high"]

    ws.append([
        variable,
        VERY_HIGH_THRESHOLD,
        info["title"],
        info["description"]
    ])


# ==========================================================
# Balance Consideration
# ==========================================================
ws = wb.create_sheet("Balance Consideration")

ws.append([
    "key",
    "title",
    "description"
])

for key, info in BALANCE_CONSIDERATION_LIBRARY.items():

    ws.append([
        key,
        info["title"],
        info["description"]
    ])

wb.save("library.xlsx")
```
