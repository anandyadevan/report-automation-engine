```python
import numpy as np
import pandas as pd

# Load enriched dataset prepared for the reporting prototype
df = pd.read_excel("student_performance_with_names.xlsx")

SUBJECT_COLUMNS = [
    "mathematics_score",
    "science_score",
    "language_score",
    "social_studies_score",
]


def generate_subject_scores(student):
    """
    Generate consistent subject scores for each student.

    The generated scores:
    - are centered around the student's total score
    - remain reproducible using student_id as the random seed
    - stay within the 0–100 score range
    """

    base_score = student["total_score"]

    rng = np.random.default_rng(int(student["student_id"]))

    variations = rng.normal(loc=0, scale=5, size=len(SUBJECT_COLUMNS))

    subject_scores = base_score + variations

    subject_scores -= subject_scores.mean()
    subject_scores += base_score

    subject_scores = np.clip(subject_scores, 0, 100)

    return pd.Series(
        subject_scores.round(1),
        index=SUBJECT_COLUMNS,
    )


df[SUBJECT_COLUMNS] = df.apply(
    generate_subject_scores,
    axis=1,
)

df.to_excel(
    "student_performance_enriched.xlsx",
    index=False,
)
```

**Note: The original Kaggle dataset was lightly enriched with synthetic student names to improve report readability. This preprocessing step is not part of the reporting architecture and is therefore omitted from this repository.**
